```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

df = pd.read_csv("CAR-JORDAN-1707561745.csv")

# Sample data
x = df.iloc[:,0]
y = df.iloc[:,1]
z = df.iloc[:,2]

x1 = df.iloc[:,3]
y1 = df.iloc[:,4]
z1 = df.iloc[:,5]

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot points for first set of coordinates
ax.scatter(x, y, z, c='r', marker='o', label='Left')

# Plot points for second set of coordinates
ax.scatter(x1, y1, z1, c='b', marker='^', label='Right')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Add a legend
ax.legend()
plt.show()
```

```python
import pandas as pd

# Load all CSV files into one DataFrame
dfs = []
for i in range(1, 11):
    filename = f"Gesture-2-{i}.csv"
    df = pd.read_csv(filename,header=None)
    dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True)
```

```python
combined_df
```

```python
# Split the DataFrame into smaller DataFrames of 40 rows each
split_dfs = [combined_df[i:i+40] for i in range(0, len(combined_df), 31)]

# Save each smaller DataFrame to a separate CSV file
for i, split_df in enumerate(split_dfs):
    split_df.to_csv(f"Gesture-2.{i+1}.csv", index=False, header=False)
```
